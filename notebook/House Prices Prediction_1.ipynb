```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv ("train.csv")

# ---ETAPE 1 : CLEANING

cols_na_to_none = ["Alley","PoolQC","Fence","MiscFeature","FireplaceQu", "GarageType"]
for col in cols_na_to_none:
    if col in df.columns:
        df[col] = df[col].fillna("None")

df["LotFrontage"] = df["LotFrontage"].fillna(df["LotFrontage"].median())


# ---ETAPE 2 : Grafik
plt.figure(figsize=(12, 5))

#Grafik1
plt.subplot(1, 2, 1)
sns.histplot(df["SalePrice"], kde=True, color="blue")
plt.title("Distribution of sales prices")

#Grafik2
plt.subplot(1, 2, 2)
sns.scatterplot(data=df, x="GrLivArea", y="SalePrice", alpha = 0.5)
plt.title("Price vs. Living Area")

plt.tight_layout()
plt.show()

# ---ETAPE 3 : ENCODAGE
le = LabelEncoder()
obj_cols = df.select_dtypes(include=["object"]).columns
for col in obj_cols:
    df[col] = le.fit_transform(df[col].astype(str))

print("Data and graphs ready.")
```

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv ("train.csv")

# ---ETAPE 1 : CLEANING

cols_na_to_none = ["Alley","PoolQC","Fence","MiscFeature","FireplaceQu", "GarageType"]
for col in cols_na:
    if col in df.columns:
        df[col] = df[col].fillna("None")


#---Etape 2 : ENCODAGE
le = LabelEncoder()
obj_cols = df.select_dtypes(include=["object"]).columns
for col in obj_cols:
    df[col] = le.fit_transform(df[col].astype(str))

#ETAPE 3: AFFICHAGE DU TABLEAU ET DES GRAPHES---
print(df.head())

#Generation des graphique

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.countplot(data=df, x="MSSubClass")
plt.title("Distribution of housing types")


plt.subplot(1, 2, 2)
sns.histplot(df["SalePrice"], kde=True, color="blue")
plt.title("Distribution of sales prices")


plt.show()


```
